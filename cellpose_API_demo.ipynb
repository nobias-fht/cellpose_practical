```python
import os
import skimage
from cellpose import models
from cellpose.plot import mask_overlay, show_segmentation
```

```python
#update these to the appropriate paths
input_path = ''
output_path = ''
model_path = ''
os.makedirs(output_path, exist_ok=True)

file_list = os.listdir(input_path)

print(file_list)
```

```python
model = models.CellposeModel(pretrained_model=model_path, gpu=True)
```

```python
for file in file_list:

    im = skimage.io.imread(os.path.join(input_path, file))
    masks, flows, styles = model.eval(im)
    skimage.io.imsave(os.path.join(output_path, file), masks)
```
